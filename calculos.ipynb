#### Imports de bibliotecas

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import statsmodels.api as sm
```

#### Carregar dados

```python
file_path = 'C:\\Codes\\trabalho-estatistica\\grupo9_projeto9.csv'
colunas_analise = ['site_id', 'mean_annual_precip', 'sand_fraction', 'soil_erosion']
try:
    df = pd.read_csv(file_path)
except FileNotFoundError:
    print(f"O arquivo não foi encontrado")

print(df.head())
```

#### Processamento dos dados

##### Média, mediana e moda

```python
for col in colunas_analise:
    if pd.api.types.is_numeric_dtype(df[col]):
        media = df[col].mean()
        mediana = df[col].median()
        moda = df[col].mode()
        moda_str = ', '.join(map(str, moda.values)) if not moda.empty else 'sem valor'
        print(f"{col} - Média: {media:.2f} | Mediana: {mediana:.2f} | Moda: {moda_str}")
    else:
        moda = df[col].mode()
        moda_str = ', '.join(map(str, moda.values)) if not moda.empty else 'sem valor'
        print(f"{col} - Moda: {moda_str}")
```
